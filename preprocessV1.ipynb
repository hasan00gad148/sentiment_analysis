```python
import os
```

```python
path_pos = "./datav1/pos.txt"
path_neg = "./datav1/neg.txt"
```

```python
word_pos=[]
word_neg=[]
```

```python
from nltk.tokenize import wordpunct_tokenize
from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
```

```python

f = open(path_pos, 'r')
lines_pos = f.readlines()
for l in lines_pos:
    lw = wordpunct_tokenize(l)     
    for w in lw:
        word_pos.append(stemmer.stem(w))
```

```python
len(word_pos)
```

```python
f = open(path_neg, 'r')
lines_neg = f.readlines()
for l in lines_neg:
    lw = wordpunct_tokenize(l)
    for w in lw:
        word_neg.append(stemmer.stem(w))
```

```python
len(word_neg)
```

```python
from nltk.corpus import stopwords # Text data
import string
```

```python
stop_words = stopwords.words("english")
stop_words.append(list(string.punctuation))
stop_words = (stop_words)
filtered_word_pos = [ word for word in word_pos if word not in stop_words and not w.isnumeric()]
filtered_word_neg = [ word for word in word_neg if word not in stop_words and not w.isnumeric()]
```

```python
n_pos = len(filtered_word_pos)
n_neg = len(filtered_word_neg)
n_pos,n_neg
```

```python
from collections import Counter
import numpy as np
```

```python
vocap = set(filtered_word_pos)
vocap2 = set(filtered_word_neg)
vocap=vocap.union(vocap2)
len(vocap)
```

```python
x_pos=Counter(filtered_word_pos)
(np.log(x_pos["fun"])-np.log(n_pos))
```

```python
x_neg=Counter(filtered_word_neg)
np.log(x_neg["fun"]/n_neg)
```

```python
x_pos["fun"]/n_pos,x_neg["fun"]/n_neg
```

```python
logprior = np.log(n_pos/n_neg)
```

```python
vocap_prob = {}
for v in vocap: 
  vocap_prob[v] = np.log((x_pos.get(v,0)+1)/(n_pos+1)) - np.log((x_neg.get(v,0)+1)/(n_neg+1)) 
```

```python
vocap_prob,logprior
```

```python
acc = 0
for i in range(0,5140):
    lw = wordpunct_tokenize(lines_pos[i])
    words = []
    for w in lw:
        words.append(stemmer.stem(w))
    filtered_words = [ word for word in words if word not in stop_words and not w.isnumeric()]
    sum = logprior
    for w in filtered_words:
        sum += vocap_prob.get(w, 0)
    acc += 1 if sum > 0 else 0
acc/5140
```

```python
acc = 0
for i in range(0,5140):
    lw = wordpunct_tokenize(lines_neg[i])
    words = []
    for w in lw:
        words.append(stemmer.stem(w))
    filtered_words = [ word for word in words if word not in stop_words and not w.isnumeric()]
    sum = logprior
    for w in filtered_words:
        sum += vocap_prob.get(w, 0)
    acc += 1 if sum < 0 else 0
acc/5140
```
